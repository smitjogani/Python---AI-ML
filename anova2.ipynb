```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
np.random.seed(12)
```

```python
world = ["asian", "black", "hispanic", "others", "white"]
```

```python
voter = np.random.choice(a = world, p = [0.05, 0.15, 0.25, 0.05, 0.5], size = 1000)
```

```python
voter_age = stats.poisson.rvs(loc = 18, mu = 30, size = 1000)
```

```python
voter_frame = pd.DataFrame({"voter":voter, "age": voter_age})
groups = voter_frame.groupby("voter").groups
```

```python
asian = voter_age[groups["asian"]]
black = voter_age[groups["black"]]
hispanic = voter_age[groups["hispanic"]]
others = voter_age[groups["others"]]
white = voter_age[groups["white"]]
```

```python
stats.f_oneway(asian, black, hispanic, others, white)
```

```python
voter_pairs = []
```

```python
for voter1 in range(4):
  for voter2 in range(voter1+1,5):
    voter_pairs.append((voter[voter1],voter[voter2]))
```

```python
for voter1, voter2 in voter_pairs:
  print(voter1, voter2)
  print(stats.ttest_ind(voter_age[groups['voter1']]))
```
